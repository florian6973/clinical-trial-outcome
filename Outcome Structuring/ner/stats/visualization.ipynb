```python
data = {}
for i, x in enumerate(top10conditions):
    df_cl = df[df['snomed_top_level_term'] == x].groupby('object')['nct_id'].count().sort_values(ascending=False).index.values[:5]
    print(x, df_cl)
    data[x] = df_cl
```

```python
outcomes = list(set([item for sublist in data.values() for item in sublist]))  # Unique outcomes
df = pd.DataFrame(0, index=outcomes, columns=data.keys())

# Assign rank values
for category, outcomes_list in data.items():
    for rank, outcome in enumerate(outcomes_list, start=1):
        df.loc[outcome, category] = rank
```

```python
import numpy as np

non_null_counts = (df != 0).sum(axis=1)

# Replace 0s with NaN for average rank calculation
average_rank = df.replace(0, np.nan).mean(axis=1)

# Combine non-null counts and average rank into a DataFrame for sorting
ordering_df = pd.DataFrame({
    'NonNullCount': non_null_counts,
    'AverageRank': average_rank
})

# Sort outcomes by non-null counts (descending) and average rank (ascending)
ordering_df = ordering_df.sort_values(by=['NonNullCount', 'AverageRank'], ascending=[False, True])
ordered_outcomes = ordering_df.index

# Reorder DataFrame based on the sorted outcomes
df_ordered = df.loc[ordered_outcomes, :]
```

```python
df_transposed = df.T

# Plot the transposed heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df_transposed, annot=True, cmap="YlGnBu", cbar_kws={'label': 'Rank'}, linewidths=.5)
plt.title("Heatmap of Categories by Measured Outcomes with Rank Values (Transposed)")
plt.xlabel("Measured Outcomes")
plt.ylabel("Category")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Plot the reordered heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df_ordered.T, annot=True, cmap="YlGnBu", cbar_kws={'label': 'Rank'}, linewidths=.5)
plt.title("Heatmap of Categories by Measured Outcomes (Ordered by Non-Null Count and Average Rank)")
plt.xlabel("Measured Outcomes")
plt.ylabel("Category")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
slopes = []

for index, row in dfg.reset_index()[:100].iterrows():
    dfi = df[(df['snomed_top_level_term'] == row['snomed_top_level_term']) & (df['object'] == row['object'])]
    dfi = dfi.dropna()
    dfi['start_date_year'] = dfi['start_month_year'].copy().apply(lambda x:x[:4])
    # dfi = dfi[(dfi['start_date_year'] > '2005') & (dfi['start_date_year'] < '2015')]
    dfi = dfi[(dfi['start_date_year'] > '2006') & (dfi['start_date_year'] < '2020')]

    # normalize counts

    # print(dfi[['start_date_year']])
    counts_per_year = dfi['start_date_year'].value_counts().sort_index()
    counts_per_year = (counts_per_year / gd).dropna()
    # print()
    # break
    counts_df = counts_per_year.reset_index()
    counts_df.columns = ['year', 'count']

    # Linear Regression
    X = counts_df[['year']]  # Independent variable (reshape for sklearn)
    y = counts_df['count']   # Dependent variable

    # Fit the linear regression model
    model = LinearRegression()
    model.fit(X, y)

    # Get the coefficient (slope)
    slope = model.coef_[0]

    # Output the slope
    # print(f"Slope of the trend (linear regression coefficient): {slope}")
    slopes.append(abs(slope))

    # Optional: Visualize

print(sorted(slopes, reverse=True))
```

```python
data_array = np.array(slopes)

# Get the indices of the 10 largest values
indices_of_top_10 = data_array.argsort()[-10:][::-1]  # Sort indices, get last 10, reverse order
indices_of_top_10
```

```python
import matplotlib.pyplot as plt
```

```python
# normalize by the number of trials outcomes in the year

plt.figure(figsize=(15, 10))
for idx in indices_of_top_10:
    row = dfg.reset_index().iloc[idx]
    dfi = df[(df['snomed_top_level_term'] == row['snomed_top_level_term']) & (df['object'] == row['object'])]
    dfi = dfi.dropna()
    dfi['start_date_year'] = dfi['start_month_year'].copy().apply(lambda x:x[:4])
    # dfi = dfi[(dfi['start_date_year'] > '2005') & (dfi['start_date_year'] < '2015')]
    dfi = dfi[(dfi['start_date_year'] > '2006') & (dfi['start_date_year'] < '2020')]

    dfigb = (dfi.groupby('start_date_year').count()['nct_id'] / gd).dropna()
    print(dfigb)
    plt.plot(dfigb.index, dfigb, label=f"{row['snomed_top_level_term']} {row['object']}")

plt.xlabel("year")
plt.ylabel("count")
plt.legend()
plt.tight_layout()
```
